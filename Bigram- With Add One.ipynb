```python
textFile = open(r"D:\MS IDS-2\LanguageModelingCorpus\RestaurantCorpusShort.txt", "r")
word_list = []
lines = []

word_list.append('<s>')
with textFile as f:
    for line in f:
        #print(line)
        lines = line
        for word in lines.split(None)[1:]:
            word_list.append(word)
        word_list.append('</s>')
        word_list.append('<s>')
        #lines.append(line)
   # b = f.read().replace("i'd", "i would")
    #word_list = b.split(' ')
word_list
```

```python
biGramProb = {}
checkWord = ''
uniqueWord = ''
for i in range(1, len(word_list)):
    checkWord = word_list[i-1]+'|'+word_list[i]
    if not checkWord in biGramProb:
        biGramProb[checkWord] = 0
    else:
        biGramProb[checkWord] += 1
    
print(biGramProb)        
```

```python
uniGramWords = {}
for word in word_list:
    uniGramWords[word] = word_list.count(word)
uniGramWords
```

```python
print(len(biGramProb))
print(len(uniGramWords))
word = ''
for key,value in biGramProb.items():
    word = key.split('|')[0]
    if word in uniGramWords:
        biGramProb[key] = value/uniGramWords[word]
print(biGramProb)
```

```python
testLine = []
test_line = 'i would like to eat chinese food'
for word in test_line.split(' '):
    testLine.append(word)
testLine.insert(0,'<s>')
testLine.append('</s>')
testLine
```

```python
checkWord = ''
totalProbability = 1
logProbability = 1
for i in range(1, len(testLine)):
    word = testLine[i-1]+'|'+testLine[i]
    if word in biGramProb:        
        totalProbability = totalProbability * biGramProb[word]
        logProbability = logProbability + math.log(biGramProb[word])
        print(word + ":" + str(biGramProb[word]))
print("Total Probability :" + str(round(totalProbability, 4)))
print("Total Log Probability :" + str(math.exp(logProbability)))
```
